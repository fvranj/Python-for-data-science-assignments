```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dinamo_hajduk = pd.read_html("https://hnl.hr/povijest/medjusobni-susreti/")
```

```python
derbi = dinamo_hajduk[0]
```

```python
derbi["Utakmica IzvjeÅ¡taj"]=pd.to_datetime(derbi["Utakmica IzvjeÅ¡taj"])
```

```python
derbi = derbi[derbi["Utakmica IzvjeÅ¡taj"]>'1-1-1995'][["Utakmica IzvjeÅ¡taj","Utakmica IzvjeÅ¡taj.1","Utakmica IzvjeÅ¡taj.2","Utakmica IzvjeÅ¡taj.4"]].dropna()
```

```python
derbi_poljud = derbi[derbi["Utakmica IzvjeÅ¡taj.1"]=="Hajduk"]
derbi_maks = derbi[derbi["Utakmica IzvjeÅ¡taj.1"]=="Dinamo"]
derbi_poljud
```

```python
derbi_maks.loc[29, 'Utakmica IzvjeÅ¡taj'] = '2015-12-05'
```

```python
derbi_poljud['goal_difference'] = derbi_poljud['Utakmica IzvjeÅ¡taj.4']-derbi_poljud['Utakmica IzvjeÅ¡taj.2']
derbi_maks['goal_difference'] = derbi_maks['Utakmica IzvjeÅ¡taj.2']-derbi_maks['Utakmica IzvjeÅ¡taj.4']
```

```python
poljud_goals = derbi_poljud.groupby("goal_difference").agg({"Utakmica IzvjeÅ¡taj.1":"count"})
```

```python
maks_goals = derbi_maks.groupby("goal_difference").agg({"Utakmica IzvjeÅ¡taj.1":"count"})
```

```python
font_color = '#525252'
hfont = {'fontname':'Calibri'}
facecolor = '#eaeaf2'
column0 = poljud_goals['Utakmica IzvjeÅ¡taj.1']
column1 = maks_goals['Utakmica IzvjeÅ¡taj.1']

fig, axes = plt.subplots(figsize=(10,5), facecolor='#eaeaf2', ncols=2, sharey=True)
fig.tight_layout()

bar_polj = axes[0].barh(poljud_goals.index, column0, align='center', color=['red','red','red','grey','blue','blue','blue'])
axes[0].set_title('Games played on Hajduk\'s home stadium', fontsize=12, pad=15)
bar_din = axes[1].barh(maks_goals.index, column1, align='center', color=['red','red','grey','blue','blue','blue','blue'])
axes[1].set_title('Games played on Dinamo\'s home stadium', fontsize=12, pad=15)
axes[0].invert_xaxis() 
plt.subplots_adjust(wspace=0, top=0.85, bottom=0.1, left=0.18, right=0.95)
axes[0].set_yticks([ -3, -2, -1, 0, 1, 2, 3, 4]);
axes[0].set_yticklabels([3,2,1,0,1,2,3,4]);
axes[0].set_ylabel("Goal difference", fontsize=14);
fig.text(0.565, 0.02, 'Number of games', ha='center', fontsize=14)
axes[1].set_xlim(0,18)
axes[0].set_xlim(18,0)
axes[0].bar_label(bar_polj, padding = -13, fontsize=10);
axes[1].bar_label(bar_din, padding = 2, fontsize=10);

# Creating legend with color box
din_vic = mpl.patches.Patch(color='blue', label='Dinamo win')
draw = mpl.patches.Patch(color='grey', label='Draw')
haj_vic = mpl.patches.Patch(color='red', label='Hajduk win')
axes[0].legend(handles=[din_vic, draw, haj_vic]);
plt.savefig('Assignment_4')
```
